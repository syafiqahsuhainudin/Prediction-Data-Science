```python
def predict(Negeri,House,Years,housesize,flooring,flooring_material,siling,Siling_material,Design_Cabinet,Jenis_Cabinet,saizbaseunit,saizwallunit,Buitin_cabinet,materia_pintukabinet,material_handle,Materialcountertop,Material_backsplash,sinki,Elektrik,Kerja_mengecat,Wiring_surface,wiring_conceal,kitchensize,Salary):
    output = np.zeros(93)
    
    output[0] = Years
    output[1] = housesize
    output[2] = flooring
    output[3] = siling
    output[4] = saizbaseunit
    output[5] = saizwallunit
    output[6] = Elektrik
    output[7] =Kerja_mengecat
    output[8] =kitchensize
  
    return model.predict([output])[0]
    
```
